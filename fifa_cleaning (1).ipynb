# **My First ML Project**

## **Loading Data**

```python
from google.colab import files
import pandas as pd
import io
print("select the dataset to upload")
uploaded = files.upload()
for filename, content in uploaded.items():
  df = pd.read_excel(io.BytesIO(content))
  print(df)
```

# **Data Cleaning**

## **checking the data**

```python
df.head()
df.info()
```

### **Removing Duplicates**

```python
df.drop_duplicates()
```

## **fixing country names**

```python
df['nationality'] = df["nationality"].replace('spanish','spain')
df['nationality'] = df["nationality"].replace('american','usa')
df['nationality'] = df["nationality"].replace('Iraqi','Iraq')
df['nationality'].unique()
df['nationality'].value_counts()

```

## **finding the best players**

```python
df.nlargest(10,'creativity_score')[["player_name","nationality","team","creativity_score"]]
```

# **feature engineering**

***overall performance score***

```python
df['overall_score']=(df['creativity_score']+df['possession_impact']+df['pressure_resistance']+df['consistency_score'])/4
print('new features created')
```

```python
df.to_csv('fifa_cleaned _with _features.csv',index=False, sep=',')
files.download('fifa_cleaned _with _features.csv')
```
